```python
df['Year'] = pd.to_datetime(df['Year'])
df.dtypes
```

```python
# Holding out test set before preforming training & validating dataset.

start_date = '2011-01-01'
end_date = '2015-01-01'

test = df[(df['Year'] > start_date) | (df['Year'] == end_date)]
test.shape
```

```python
from sklearn.model_selection import train_test_split

train, val = train_test_split(df, train_size= .50)
```

```python
features = ['NA_Sales', 'JP_Sales', 'EU_Sales']
```

```python
target = 'Global_Sales'

# Getting a dataframe with all train columns except target
X_train = train.drop(columns=[target])

# Getting a list of numeric features
numeric_features = train.select_dtypes(include='number').columns.tolist()

# Getting a series with the cardinality of nonnumeric features

cardinality = train.select_dtypes(exclude='number').nunique()

# Get a list of all categorical features with cardinality <= 50
categorical_features = cardinality[cardinality <= 50].index.tolist()

# Combining the list

features = numeric_features + categorical_features

print(features)
```

```python
# Arranging data into X features matrix and y Target vectors

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
```

```python
from sklearn.preprocessing import OneHotEncoder

X_train = OneHotEncoder.fit_transform(X_train)
X_test = OneHotEncoder.transform(X_test)

```

```python
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier


pipeline = make_pipeline(
    OneHotEncoder(),
    SimpleImputer(strategy='median'),
    RandomForestClassifier(random_state=20, n_jobs=-1)
)



print('Validation Accuracy', pipeline.score(X_val, y_val))
print('Train Accuracy', pipline.score(X_train, y_train))

y_pred = pipeline.predict(X_val)
```
